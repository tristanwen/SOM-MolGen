```python
import numpy as np
import pandas as pd
from test_fp_similarity import cal_fingerprint
import codes_for_manuscript_02.Regressors_train_retrain_opt_test as my_MLs
from codes_for_manuscript_02.list_hypopt import sklearn_machines_hypopt_list as hypist
```

```python
## 准备工作

# 导入数据
root = "F:\\WORK\\和外合作\\天宇媛媛毕设\\QSAR\\"
raw_dataset = pd.read_excel(root+"pIGC50.xlsx")
dataset = raw_dataset[['pIGC50/exp', 'MW', 'nDB', 'GATS3m',
                       'GATS1p','TDB03u', 'Rm3s', 'nROH', 'H046', 'B02[N-O]', 'MLOGP2']]
print(dataset.head())
# 训练数据 yX
all_data = dataset.values
```

```python
# 计算fp
run_fp_cal = 0
if run_fp_cal:
    fp_table1 = cal_fingerprint(dataset["Smiles"].tolist())
    fp_table1.RDKitTopological(fpSize=1024)
    fp_table2 = fp_table1.fp_TopoFingerprint

    # 保存fp
    fp_table2.to_csv("F:\\WORK\\和外合作\\天宇媛媛毕设\\QSAR\\RDKitTopo_fp.csv",index=False)

fp_table2 = pd.read_csv("F:\\WORK\\和外合作\\天宇媛媛毕设\\QSAR\\RDKitTopo_fp.csv", index_col=None)
all_data = fp_table2.values
fp_table2.head()


```

```python
## 超参数优化

model_name = '随机森林回归'# '神经网络','随机森林回归','支持向量机'
model_save = root
res2 = {} # 保存测试集上的评价指标

hypopt_MP_opt = hypist(choose_machines=()).hypopt_list # 该模型的超参数字典

a = hypopt_MP_opt[0][model_name],hypopt_MP_opt[1][model_name]

training = my_MLs.training_machine(model_name=model_name, all_data=all_data,trial=False,
                                   data_loading_mode=(0.3, 2023, "train"),
                                   train_mode="hypopt", model_save=model_save,
                                   hypopt_MP_opt=a,hypopt_cv=5,hypopt_max_evals=100)
print("---------hypopt train--------:\n---------training result of %s----------\n" % model_name)
res1 = {i: training.overall_metrics[i] for i in ['可解释方差得分evs', '均方根误差rmse', '均方差mse']}
print(res1)

testing = my_MLs.testing_machine(trained_model=training.final_model,model_name=model_name,all_data=all_data,trial=False,
                                 data_loading_mode=(0.3, 2023, "test"),result_save=model_save)
print("---------hypopt train--------:\n---------testing result of %s----------\n"%model_name)
res2[model_name] = {i: testing.overall_metrics[i] for i in ['可解释方差得分evs', '均方根误差rmse', '均方差mse']}
print(res2[model_name])
```
